```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
roc = pd.read_csv('../data/FDDB/temp/DiscROC_detr.txt', sep=' ', header=None)
roc.columns = ['tpr', 'fp', 'threshold']
roc_fb = pd.read_csv('../data/FDDB/temp/DiscROC_fb.txt', sep=' ', header=None)
roc_fb.columns = ['tpr', 'fp', 'threshold']
roc_mt = pd.read_csv('../data/FDDB/temp/DiscROC_mtcnn.txt', sep=' ', header=None)
roc_mt.columns = ['tpr', 'fp', 'threshold']
```

```python
def plot_roc():
    _, axis = plt.subplots(nrows=1, ncols=1, figsize=(7, 4), dpi=120)
    axis.plot(roc.fp, roc.tpr, c='r', linewidth=2.0,label='DETR');
    axis.plot(roc_fb.fp, roc_fb.tpr, c='g', linewidth=2.0,label='FaceBoxes');
    axis.plot(roc_mt.fp, roc_mt.tpr, c='b', linewidth=2.0,label='MTCNN');
    axis.set_title('Discontinuous ROC curves')
    axis.set_xlim([0, 1000.0])
    axis.set_ylim([0.3, 1.0])
    axis.set_xlabel('False Positives')
    axis.set_ylabel('True Positive Rate');
    axis.legend()
    plt.savefig('DISC_ROC.png')

plot_roc()
```

```python
roc = pd.read_csv('../data/FDDB/temp/ContROC_detr.txt', sep=' ', header=None)
roc.columns = ['tpr', 'fp']
roc_fb = pd.read_csv('../data/FDDB/temp/tempContROC_faceboxes.txt', sep=' ', header=None)
roc_fb.columns = ['tpr', 'fp']
roc_mt = pd.read_csv('../data/FDDB/temp/ContROC_mtcnn.txt', sep=' ', header=None)
roc_mt.columns = ['tpr', 'fp']
```

```python
def plot_roc():
    _, axis = plt.subplots(nrows=1, ncols=1, figsize=(7, 4), dpi=120)
    axis.plot(roc.fp, roc.tpr, c='r', linewidth=2.0,label='DETR');
    axis.plot(roc_fb.fp, roc_fb.tpr, c='g', linewidth=2.0,label='FaceBoxes');
    axis.plot(roc_mt.fp, roc_mt.tpr, c='b', linewidth=2.0,label='MTCNN');
    axis.set_title('Continuous ROC curves')
    axis.set_xlim([0, 1000.0])
    axis.set_ylim([0.0, 1.0])
    axis.set_xlabel('False Positives')
    axis.set_ylabel('True Positive Rate');
    axis.legend()
    plt.savefig('CONT_ROC.png')

plot_roc()
```
